```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
%matplotlib inline
while True:
    try:
        params = [int(i) for i in input("Input the coefficient of a third-order equation: (format a, b, c, d)\n").strip().split(',')]
       
        if len(params) == 4:
            k = 0
            i = 0            
            for i in range(4):
                if params[i] >=0 or params[i] < 0:
                    
                    k+=1
                    continue
                else:    
                    break
            if k < 3 or ((k == 3) and not(params[k] >= 0 or params[k] < 0)):
                continue
            else:
                break
        else:
            continue
        
        
    except:
        print("Wrong Input!")
def f(x, a, b, c, d):
    return a*x**3 + b*x**2 + c*x + d

def draw_figure():
    X = np.linspace(-100,100,100)
    Y = f(X, params[0], params[1], params[2], params[3])
    plt.ylabel("Y")
    plt.xlabel("X")
    plt.plot(X, Y,"k-")
    
draw_figure()
plt.title("HW1")

while True:
    dots = [int(i) for i in input("Input the X , Y\n").strip().split(",")]
    if len(dots)%2 == 0:
        break
list_x = []
list_y = []
for i in range(len(dots)):
    if i%2 == 0:
        list_x.append(dots[i])
    else:
        list_y.append(dots[i])
        
for i in range(len(list_x)):
    y = params[0]*list_x[i]**3 + params[1]*list_x[i]**2 + params[2]*list_x[i] + params[3]
    if list_y[i] > y:
        plt.plot(list_x[i], list_y[i], "ro")
    elif list_y[i] < y:
        plt.plot(list_x[i], list_y[i], "bo")
    else:    
        plt.plot(list_x[i], list_y[i], "go")
        
plt.show()
plt.clf()


max_iter = 10
eta = 0.1


def gradient_figure(pos_neg):
    x = -10
    v = 0
  
    draw_figure()
    for k in range(max_iter):
        dx = 3*params[0]*x**2 + 2*params[1]*x + params[2]        
            # v表示x要改變的幅度
        v =  float(pos_neg) * dx * eta 
       
        y = f(x, params[0], params[1], params[2], params[3]) 
        plt.plot(x, y, "go")
        x += v 
  
        
    y = f(x, params[0], params[1], params[2], params[3])
    plt.plot(x, y, "ro")
    plt.show()
    plt.clf()
"""find the biggest value"""
gradient_figure(1)     
"""find the smallest value"""
gradient_figure(-1)    

```
